Розгляньте групу матриць за множенням  

$$G = \left\langle \left(\begin{array}{rrr}
1 & 0 & 1 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{array}\right),  \left(\begin{array}{rrr}
1 & 0 & 0 \\
0 & 1 & 1 \\
0 & 0 & 1
\end{array}\right)\right\rangle$$

Дайте відповідь на наступні питання: 

1. _Чи є група G скінченною?_
2. _Чи є група G комутативною?_
3. _Який вигляд мають елементи групи G?_
4. _Які порядки мають елементи групи G?_ 
5. _Чому ізоморфна група G?_


__Відповіді__: 
1.  Група нескінченна, бо матриці генерують нескінченно багато різних елементів (якщо наприклад підносити першу матрицю до степеня n ∈ ℤ,  елемент (1,3) утвореної матриці дорівнюватиме n, відповідно таких матриць нескінченна кількість)  
2.  Так, для довільних A та B виконується рівність A * B = B * A
3.  Оскільки група комутативна, кожен її елемент можна подати у вигляді A ** n * B ** m = [[1, 0, n], [0, 1, m], [0, 0, 1]], де n,m довільне ціле число
4. Порядок у всіх нескінченний 
5. Група ізоморфна декартовому добутку на множині цілих чисел, бо кожен її елемент можна подати парою (n,m)

```python
A = matrix([[1, 0, 1], [0, 1, 0], [0, 0, 1]])
B = matrix([[1, 0, 0], [0, 1, 1], [0, 0, 1]])
B ** -6 * A ** -5
```

```python
r = A * B ** -1 * B ** 3 * A ** -3
l = B ** -1 * A * A ** -3 * B ** 3
r, l
```
